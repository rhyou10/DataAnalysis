# 데이터 프레임 합성

```python
import numpy as np
import pandas as pd
```

## merge 함수를 사용한 데이터프레임 병합

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
# inner join 공통부분
pd.merge(df1,df2)
```

```python
#left outer join 왼쪽기준으로
pd.merge(df1,df2, how='left')
```

```python
#right outer join 오른쪽 기준
pd.merge(df1,df2, how='right')
```

```python
# outer join 키값이 하나만 있어도 합침
pd.merge(df1,df2, how='outer')
```

```python
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1
```

### on

```python
df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
df2
```

```python
pd.merge(df1, df2, on='고객명')
```

```python
df1 = pd.DataFrame({
    '이름': ['영희', '철수', '철수'],
    '성적': [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({
    '성명': ['영희', '영희', '철수'],
    '성적2': [4, 5, 6]})
df2
```

```python
pd.merge(df1, df2, left_on='이름', right_on="성명")
```

```python
df1 = pd.DataFrame({
    '도시': ['서울', '서울', '서울', '부산', '부산'],
    '연도': [2000, 2005, 2010, 2000, 2005],
    '인구': [9853972, 9762546, 9631482, 3655437, 3512547]})
df1
```

### join 메소드
할떄 인덱스를 같게해야한다

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1.set_index('고객번호', inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2.set_index('고객번호', inplace=True)
df2
```

```python
df1.join(df2)
```

### concat
그냥 합치는거

```python
s1 = pd.Series([0, 1], index=['A', 'B'])
s1
```

```python
s2 = pd.Series([2, 3, 4], index=['A', 'B', 'C'])
s2
```

```python
pd.concat([s1, s2])# 아래로 붙인다 리스트형태로 넣어야한다
```

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['데이터1', '데이터2'])
df1
```

```python
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2, 2),
    index=['a', 'c'],
    columns=['데이터3', '데이터4'])
df2
```

```python
pd.concat([df1, df2], axis=1)
```

### 연습문제

```python
np.random.seed(2021)
df1 = pd.DataFrame(np.random.randint(100,999,size=(6,2)),
                    columns=['매출', '비용'],
                    index = ['1','2','3','4','5','6'])
df1['이익'] = df1.매출 - df1.비용

df1
```

```python
np.random.seed(2020)
df2 = pd.DataFrame(np.random.randint(100,999,size=(6,2)),
                    columns=['매출', '비용'],
                    index = ['7','8','9','10','11','12'])

df2['이익'] = df2.매출 - df2.비용
df2
```

```python
df = pd.concat([df1, df2])
df
```

```python
tdf = pd.DataFrame(df.sum()).T
tdf
```

```python
tdf.index = ['합계']
tdf
```

```python
pd.concat([df,tdf])
```
