## apply 변환

>행 또는 열에 함수를 넣고 싶을때 apply 인수에 함수를 넣으면 함수실행된다

```python
df3 = pd.DataFrame({
    'A': [1, 3, 4, 3, 4],
    'B': [2, 3, 1, 2, 3],
    'C': [1, 5, 2, 4, 4]
})
df3
```

```python
df3.apply(lambda x: x.max()-x.min())
```

```python
df3.apply(lambda x: x.max()-x.min(), axis=1)
```

```python
titanic['성년'] = titanic.apply(lambda r: '성년' if r.age>=20 else '미성년', axis=1)
titanic.tail()
```

```python
titanic['category'] = titanic.apply(lambda r: r.sex if r.age>=20 else 'child', axis=1)
titanic.tail()
```

## fillna method

> NaN 값은 fillna 메서드를 사용하여 원하는 값으로 바꿀 수 있다

```python
df3
```

```python
df3.apply(pd.value_counts)
```

```python
df3.apply(pd.value_counts).fillna(0.0)
```

```python
titanic.age.fillna(titanic.age.mean(), inplace=True) 
#inplace는 고친결과가 화면에 한번 쓰이게 만들지 말고 타이타닉에 고치겠다
#기본값은 inplace =False
titanic.tail()
```

## astype() method

## 실수값을 카데고리 값으로 변환

```python
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]
```

```python
bins =[1, 20, 30, 50, 70, 100] #기준
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(ages, bins, labels=labels)
cats
```

```python
cats.codes
```

```python
np.random.seed(2021)
data = np.random.randn(1000)
cats = pd.qcut(data, 4, labels=['Q1','Q2', 'Q3', 'Q4']) 
#구간을  정확인 4단계로 나눈다 알아서(250개씩)
cats
```
