# 데이터 프레임 인덱스조작

```python
import numpy as np
import pandas as pd
```

## 인덱스 설절 및 제거

```python
np.vstack([list('ABCDE'),np.round(np.random.rand(3, 5), 2)])
```

```python
np.random.seed(2021)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3, 5), 2)]).T,
                   columns=["C1", "C2", "C3", "C4"])
df1
```

```python
df2 = df1.set_index('C1') #c1이 인덱스가 되어라
df2
```

```python
df2.reset_index()# 바꾼걸 df2 적용하고 싶으면 inplace=True
```

## 연문 4.5.1
5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 다음과 같이 만든다.

```python
np.random.seed(2021)
data = np.random.randint(40,100,size=(5,3))
df_score1 = pd.DataFrame(data, columns=['국어', '영어', '수학'])
df_score1

##학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 df_score1 을 생성한 후
```

```python
df_score1.index = ['Ailee', 'Elizabeth', 'James', 'Maria', 'Brian']
df_score1
## df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다.
```

```python
#reset_index 명령으로 이 인덱스 열을 명령으로 일반 데이터열로 바꾸여 
# 데이터프레임 df_score2을 만든다.

df_score2=df_score1.reset_index()
df_score1.index.name='이름'
df_score2

```

```python
#df_score2 = df_score2.set_index('이름')
df_score2.set_index('이름', inplace=True)
df_score2
```
