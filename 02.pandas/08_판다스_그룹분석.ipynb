## 그룹분석

### group by
뒤에 메소드 써야한다

```python
iris = sns.load_dataset('iris')
iris.groupby(iris.species).mean()
```

```python
iris.groupby(iris.species).agg(['mean', 'std'])
```

```python

df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

```python
groups = df2.groupby(df2.key1)
groups
```

```python
df2.groupby(df2.key1).sum()
```

```python
df2.groupby(df2.key1)['data2'].sum()
```

```python
df2.data2.groupby(df2.key1).sum()
```

### agg 메소드 함수 적용

```python
def peak_to_peak_ratio(x):
    return x.max() / x.min()
```

```python
iris.groupby(iris.species).agg(peak_to_peak_ratio)
```

```python
iris.groupby(iris.species).agg(lambda x : x.max()/x.min())
```

```python
iris.groupby(iris.species).apply(lambda x : x.max()/x.min())
```

```python
iris.describe()
```

```python
iris.groupby(iris.species).describe().T
```

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
mpg['manufacturer'] = mpg.name.apply(lambda x: x.split()[0])
```

```python
mpg['model'] = mpg.name.apply(lambda x: ' '.join(x.split()[1:])) #join 안하면 배열형태
```

```python
mpg.head()
```

### transform

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=["소", "중", "대"]).astype(str)


iris["petal_length_class"] = iris.groupby(iris.species).petal_length.transform(q3cut)
iris[["petal_length", "petal_length_class"]].tail(10)
```

### pivot_table 
> groupby, pivot_table 결과는 비슷하다

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
df1.pivot("도시", "연도", "인구")#행 , 열, 값
```

```python
df1.pivot_table('인구','도시','연도')
```

```python
df1.pivot_table('인구','도시','연도', margins=True, margins_name='합계') ##합계추가 margin
```

```python
df1.pivot_table('인구',['지역', '도시'],'연도', margins=True, margins_name='합계') ##합계추가 margin
```

### tips dataset 사례

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = np.round(tips.tip / tips.total_bill *100, 2)
tips.tail()
```

```python
tips.describe()
```

```python
tips.groupby('sex').count()
```

```python
tips.groupby("sex").size()
```

```python
## 흡연여부에따른 팁 비율
tips.groupby(["smoker"])[['tip_pct']].mean()
```

```python
#성별과 흡연여부에 따른 팁 비율
tips.groupby(["sex", "smoker"])[['tip_pct']].mean()
```

```python
tips.pivot_table("tip_pct", ["sex","smoker"])
```

```python
tips.pivot_table("tip_pct", "sex","smoker")
```

### 연슴문제 4.7.3

```python
#요일
tips.groupby('day')[['tip_pct']].mean()
```

```python
## 점심 저녁여부
tips.groupby('time')[['tip_pct']].mean()
```

```python
# 요일 점심 여부 동시에
tips.groupby(['day', 'time'])[['tip_pct']].mean()
```

```python
tips.pivot_table('tip_pct', 'time', 'day')
```

```python
tips.pivot_table('tip_pct', ['time', 'day'])
```

```python
# 요일 시간 인원수
tips.pivot_table('tip_pct', ['time', 'size'], 'day')
```

```python
# 어떤 용인이 가장 크게 작용?
tips.groupby(['day', 'time'])[['tip_pct']].agg(lambda x : x.max()/ x.min())
```

```python
# 어떤 용인이 가장 크게 작용?
tips.groupby(['sex', 'smoker'])[['tip_pct']].agg(lambda x : x.max()/ x.min())
```

```python
def p2p(x):
    return x.max()/x.min()
tips.groupby(['sex', 'smoker'])[['tip_pct']].agg(['mean',p2p]) ## 람다함수는 이름설정 불가능
```

```python
tips.groupby(['sex', 'smoker'])[['tip_pct']].agg(['mean','std',p2p])
```
