```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import folium
import numpy as np
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
import re
import seaborn as sns
```

```python
import json
with open('data/seoul-gu-simple.json', 'rt', encoding='UTF8') as json_file:
    geo_data = json.load(json_file)
```

```python
df = pd.read_csv('data/자치구별 공원 현황.csv')
df.head()
```

```python
center = [37.581, 126.986]      # 서울 중심부 위도, 경도
map = folium.Map(location=center, zoom_start=11, tiles='Stamen Toner')

folium.Choropleth(
    geo_data=geo_data,
    data=df,
    columns=('지역','면적'),
    key_on='feature.id',
    fill_color='BuPu',
    legend_name='면적'
).add_to(map)
title = '<h3 align="center" style="font-size:20px">서울시 자치구별 공원면적</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```
