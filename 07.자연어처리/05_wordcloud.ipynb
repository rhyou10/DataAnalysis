```python
from PIL import Image
import numpy as np
```

## 1. Alcie in wonderland

```python
textfile = 'data/Alice.txt'
maskfile = 'data/Alice_mask.png'
```

```python
with open(textfile) as fp:
    text = fp.read()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) ##영어로 스탑워드설정
stop_words.add('said')
```

```python
wc =WordCloud(background_color='white',
                max_words=1000,stopwords=stop_words)
wc.generate(text)
plt.figure(figsize=(10,16))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
keys = list(wc.words_.keys())  #워드클라우드에 들어간 단어들은 딕셔너리형태
values = list(wc.words_.values())
for i in range(10):
    print(keys[i], ':', values[i]) #가장많이나온 단어가 1
```

### 그림위에 표시기

```python
mask = np.array(Image.open(maskfile))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear') #intert 는 보강법
plt.axis('off')
plt.show()
```

```python
wc =WordCloud(background_color='white', mask=mask,
                max_words=1000,stopwords=stop_words)
wc.generate(text)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## 2.star Woars

```python
textfile = 'data/A_new_hope.txt'
maskfile = 'data/Stormtrooper_mask.png'
```

```python
with open(textfile) as fp:
    text = fp.read()
text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", 'Luke')
```

```python
from wordcloud import STOPWORDS
stop_words = set(STOPWORDS)
stop_words.add('int')
stop_words.add('ext')
len(stop_words)
```

```python
mask = np.array(Image.open(maskfile))
```

```python
wc =WordCloud(mask=mask, max_words=1000,stopwords=stop_words,
                    margin=10, random_state=2021)
wc.generate(text)
default_colors = wc.to_array
```

```python
import random
def grey_color(word, font_size, position, orientation, random_state=None, **kwargs):
    return f'hsl(0, 0%,{random.randint(60,100)}%)'
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wc.recolor(color_func=grey_color, random_state=2021),interpolation='bilinear')
plt.axis('off')
plt.show()
```
