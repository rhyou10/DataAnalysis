# 라인플롯

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# minus 표시 
# 한글 쓸때 필수다
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
```

## 라인플롯

```python
plt.title('라인플롯')
plt.plot([1, 4, 9, 16]) # y = (x + 1) ** 2
plt.grid() # 격자 선
plt.show()
## %matplotlib inline - 주피터에서 과거에 그래프를 그리기 위해서 쓰임
```

```python
plt.title('라인플롯')
xs = np.linspace(0, 3, 301) #0 3까지 301등분
y = (xs +1 ) ** 2
plt.plot(xs, y) # y = (x + 1) ** 2 
#plot (x,y) 리스트로 들어가고 하나의 값만 들어가면 y값으로
plt.grid() # 격자 선
plt.show()
```

## 스타일 지정
> [matplot](https://matplotlib.org/) 스타일 지정 

```python
plt.title("'rs--' 스타일의 plot ")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--') 
plt.grid()
plt.show()
# rs-- (r) 은 색깔 / (s) 는 마커(점 위치 표시방법) / (--)는 선표시 방법
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.grid()
plt.show()
```

## 그림 법위 지정

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50) # 보여주는 x 축 지정
plt.ylim(-2, 30) # 보여주는 y 축 지정
plt.grid()
plt.show()
```

## 틱설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]) # 축의 기준점
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

### 수학 문자식 넣기

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

## 여러개 선 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

## 겹쳐그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')# 여러개 이런식으로 그리는게 가독성 좋음
plt.grid()
plt.show()
```

## 범례(legend)
> 여러개의 라인 플롯을 동시에 그리는 경우에는 각 선이 무슨 자료를 표시하는지를 보여주기 위해

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, label="cosine") # 라벨이 있어야 범례 지정가능
plt.plot(X, S, label="sine")
plt.legend(loc='lower center') #범례 자동으로 설정 loc 를 통해 위치 설정 가능
plt.grid()
plt.show()
```

## x축, y축 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")  
plt.xlabel("time") #
plt.ylabel("amplitude") #
plt.title("Cosine Plot")
plt.grid()
plt.show()
```
