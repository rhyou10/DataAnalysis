# 판다스데이터 시각화
- plot() 을 쓸수 있다

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

# minus 표시 
# 한글 쓸때 필수다
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

### 예시

```python
np.random.seed(2021)
df1 = pd.DataFrame(np.random.randn(100, 3), #표준 정규분포 100개 3등분
                   index=pd.date_range('1/1/2021', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()## 가장 기본적인 라인플롯으로 나온다
plt.title('Pandas plot() 메소드 사용예')
plt.xlabel('일자')
plt.ylabel('DATA')
plt.grid()
plt.show()
```

## Bar chart

```python
iris.sepal_length[:20].plot(kind='bar', rot=0, 
                        color = plt.get_cmap('hot')(np.linspace(0.8, 0.2, 20))) ##kind 통해서 차트 종류 선택
                        #색깍 지정
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].plot(kind='barh', rot=0, 
                        color = plt.get_cmap('RdBu')(np.linspace(0.8, 0.2, 20))) ##kind 통해서 차트 종류 선택
                        #색깍 지정
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")
plt.xlabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].sort_values().plot(kind='barh', rot=0,  #sorting
                        color = plt.get_cmap('RdBu')(np.linspace(0.8, 0.2, 20))) ##kind 통해서 차트 종류 선택
                        #색깍 지정
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")
plt.xlabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.ylim(0,7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title("각종의 feature 별 평균")
plt.ylim(0, 8)
plt.show()
```

```python
df2.T.plot.bar(rot=0)
plt.title("각feature의 종 별 평균")
plt.ylim(0, 8)
plt.show()
```

## pie chart

```python
df3 = titanic['class'].value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객수 비용')
plt.axis('equal') #파이차트일때
plt.show()
```

## Histogram

```python
iris.plot.hist()
plt.title("각 Feature 값들의 빈도수 Histogram")
plt.xlabel("데이터 값")
plt.show()
```

## Box Plot

```python
iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box Plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
iris.boxplot(by = 'species') #알아서 타이틀까지 넣는다
plt.tight_layout(pad=1, h_pad=1)
plt.show()
```
