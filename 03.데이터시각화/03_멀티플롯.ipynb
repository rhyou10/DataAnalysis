## Axes 객체와 subplot

```python
#subplot(2, 1, 1) # 2 x 1 그래프 그리고 1째에 그리겠다
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
ax1 = plt.subplot(2, 1, 1) #2x1 그래프 1번쨰에 그리겠다 Axes
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.grid()

ax2 = plt.subplot(2, 1, 2) # Axes
plt.plot(x2, y2, 'r.-') 
plt.xlabel('time (s)')
plt.grid()
plt.ylabel('Undamped')

plt.tight_layout() #
plt.show()
```

### 반복문형태로 표시하기

```python
fig, axes = plt.subplots(2, 2) # 2 x 2 array

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 0].grid()
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2) # 2 x 2 array
np.random.seed(2021)
for i in range(4):
    axes[i//2, i%2].plot(np.random.rand(5))
    axes[i//2, i%2].set_title(f'axes{i+1}')
    axes[i//2, i%2].grid()

plt.tight_layout() # 글자 안곂치게
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(6,4)) # 2 x 2 array
np.random.seed(2021)
for i in range(4):
    ax = axes[i//2, i%2]
    ax.plot(np.random.rand(5))
    ax.set_title(f'axes{i+1}')
    ax.set_xlabel('X축'); ax.set_ylabel('Y축')
    ax.grid()
    
fig.suptitle('4개', fontsize=14)

plt.tight_layout() # 글자 안곂치게
plt.show()
```
